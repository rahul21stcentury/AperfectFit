```python
import os
import pandas as pd
import numpy as np
import re
```

```python
datadic={}
```

```python
cwd = os.getcwd()
train_path=cwd+'\\train_txt'
test_path=cwd+'\\test_txt'
resume_path=cwd+'\\dataset'
```

```python
path=train_path
files=os.listdir(path)
```

```python
###Reading 
for j in files:
    txt=open(path+'\\'+j).read()
    txt
    import re
    txt=re.split(r'[;,\s,.,),(]\s*',txt)
    txt

    import nltk
    from nltk.corpus import stopwords
    sw=stopwords.words('english')
    important=[]
    for i in txt:
        if(i.lower() not in sw):
            if(i=='' or i==' '):
                pass
            else:
                important.append(i.lower())

    one_word=important
    two_word=[important[i]+' '+important[i+1] for i in range(len(important)-1)]

    three_word=[important[i]+' '+important[i+1]+' '+important[i+2] for i in range(len(important)-2)]

    for i in one_word:
        if(datadic.get(i)==None):
            datadic[i]=1
        else:
            datadic[i]+=1
    for i in two_word:
        if(datadic.get(i)==None):
            datadic[i]=1
        else:
            datadic[i]+=1
    for i in three_word:
        if(datadic.get(i)==None):
            datadic[i]=1
        else:
            datadic[i]+=1
```

```python
import pandas as pd
words=datadic.keys()
freq=datadic.values()
table=pd.DataFrame({'Word':words,'freq':freq})
#table['freq']=table['freq']/sum(table['freq'])
```

```python
table.sort_values(['freq'],ascending=False)#.to_csv('tmp.csv',index=False)
```

```python
tp=table[table['freq']>5]
```

```python
words=list(tp['Word'])
freq=list(tp['freq'])
n=len(words)
word_filter=[]
freq_filter=[]
```

```python
for i in range(n):
    curr=None
    curr_freq=None
    remove=[]
    for j in range(i+1,n):
        flag=0
        if(words[i] in words[j]):
            for k in word_filter:
                if(k in words[i]):
                    flag=1
            if(flag==1):
                pass
            elif(freq[i]>freq[j]):
                curr_freq=freq[i]
                curr=words[i]
                remove.append(words[j])
            else:
                curr_freq=freq[j]
                curr=words[j]
    if(curr):
        word_filter.append(curr)
        freq_filter.append(freq)
```

```python
word_filter
finalwords=[]
for i in word_filter:
    if(not i.isnumeric()):
        finalwords.append(i)
```

```python
features=[x for x in finalwords if(len(x)>1)]
```

```python
k=0
for i in train_files:
    print('-----------------------------------')
    txt=open(train_path+'\\'+i+'.txt').read()
    import re
    txt=re.split(r'[;,\s,.,),(]\s*',txt)
    words=[x.lower() for x in txt]
    #print(txt)
    for j in features:
        if(j in words):
            if(word_score.get(j)==None):
                word_score[j]=[1,train_score[k]]
            else:
                word_score[j][0]+=1
                word_score[j][1]+=train_score[k]
    k+=1
```
